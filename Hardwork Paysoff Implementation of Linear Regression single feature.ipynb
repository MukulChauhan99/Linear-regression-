```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

X=pd.read_csv("Linear_X_Train.csv")
Y=pd.read_csv("Linear_Y_Train.csv")

X=X.values
Y=Y.values
```

```python
plt.style.use("seaborn")
plt.title("Marks V/S Time")
plt.xlabel("Hardwork")
plt.ylabel("Performance")
plt.scatter(X,Y)
```

```python
##Linear Regression 

def hypothesis(x,theta):
    y_=theta[0]+theta[1]*x
    return y_
    
    
def gradient(theta,X,Y):
    m=X.shape[0] ##Total Number of samples
    grad=np.zeros((2,))
    
    for i in range(m):
        
        y_=hypothesis(X[i],theta)
        grad[0]+=y_-Y[i]
        grad[1]+=(y_-Y[i])*X[i]
        
    return grad/m    
            
    
    
    
    
```

```python
def error(X,Y,theta):
    
    m=X.shape[0]
    J_=0.0
    for i in range(m):
        y_=hypothesis(X[i],theta)
        J_+=(y_-Y[i])**2
        
    return J_/m
```

```python
def gradientDescent(X,Y,learning_rate=0.1):
    theta=np.zeros((2,))
    errorlist=[]
    
    for i in range(200):    
        grad=gradient(theta,X,Y)
        e=error(X,Y,theta)
        errorlist.append(e)    
        theta[0]=theta[0]-learning_rate*grad[0]    
        theta[1]=theta[1]-learning_rate*grad[1]
    
    return theta,errorlist   
    
    
    
    
```

```python
theta,error_list=gradientDescent(X,Y)

plt.plot(error_list)
```

```python
y_=hypothesis(X,theta)
plt.style.use("seaborn")
plt.scatter(X,Y)
plt.plot(X,y_,color="orange",label="Prediction")
plt.legend()
```
